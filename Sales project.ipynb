```python
## tutorial example
import os


files = [file for file in os.listdir(r"C:\Users\Antonio\Desktop\SalesProject\2nd way merge")]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv(file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```
